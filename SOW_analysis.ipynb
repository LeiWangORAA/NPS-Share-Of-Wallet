```python
df_raw3 = data_pull('2023-01-01','2023-06-30', 'MOSSASWRKFORCE10', 'MOSwrkforce02') # We can import 6 months of data and then subset for June23 only. Or, we can extract June23 data here
print(df_raw3.shape)
df_raw3.head()
```

```python
# Some data management steps

df_raw3['SOW_match'] = df_raw3['AA_FLT_REV'].notna().astype(int) # old indicator with SOW non-missing
dt1 = df_raw3.copy()

# calculate the SOW percentage = AA/all airline spend
dt1 = dt1.groupby('NPS_CATEGORY').apply(lambda x: x.assign(sum_AA=x['AA_FLT_REV'].sum(), sum_AA_OA=x['EST_AIRLN_SPEND'].sum(), Percent_AA_to_All= 100* x['AA_FLT_REV'].sum() / (x['EST_AIRLN_SPEND'].sum() if x['EST_AIRLN_SPEND'].sum() != 0 else 1e-10))).reset_index(drop=True).drop(columns=['sum_AA', 'sum_AA_OA'])

# AA_FLT_REV is highly skewed. Create log10 for checking distribution. Hard code 0 to 0
dt1['AA_FLT_REV_log'] = np.where(dt1['AA_FLT_REV'] == 0, 0, np.log10(dt1['AA_FLT_REV']))
dt1['EST_AIRLN_SPEND_log'] = np.log10(dt1['EST_AIRLN_SPEND'])


# categorize the continuous variables
def categorize_values(col):
    conditions = [col == 0,            # Condition for "0"
                  col > 0,             # Condition for ">0"
                  col.isna() ]         # Condition for "Missing"
    choices = ["0", ">0", "Missing"]  # Corresponding categories
    return np.select(conditions, choices, default="Checking")  # Default for values < 0

# Apply the function to both columns
dt1['AA_FLT_REV_Category'] = categorize_values(dt1['AA_FLT_REV'])
dt1['AA_FLT_REV_Y1_Category'] = categorize_values(dt1['AA_FLT_REV_1stYr'])

dt1['SEG_DEP_DT'] = pd.to_datetime(dt1['SEG_DEP_DT'])

# Create the cross table
pd.crosstab(dt1['AA_FLT_REV_Y1_Category'], dt1['AA_FLT_REV_Category'], 
                          margins=True, margins_name='Total')
```

```python
# Subset data

# SOW first year not missing
dt2 = dt1[dt1['AA_FLT_REV_1stYr'].notna()]
# limit CI total_ICC
dt2 = dt2[dt2['TOTAL_ICC'].notna()]

dt2 = dt2.groupby('NPS_CATEGORY').apply(lambda x: x.assign(sum_AA=x['AA_FLT_REV'].sum(), sum_AA_OA=x['EST_AIRLN_SPEND'].sum(), 
    Percent_AA_to_All= 100* x['AA_FLT_REV'].sum() / (x['EST_AIRLN_SPEND'].sum() if x['EST_AIRLN_SPEND'].sum() != 0 else 1e-10))).reset_index(drop=True).drop(columns=['sum_AA', 'sum_AA_OA'])
dt2 = dt2.groupby('NPS_CATEGORY').apply(lambda x: x.assign(sum_AA=x['AA_FLT_REV_1stYr'].sum(), sum_AA_OA=x['EST_AIRLN_SPEND_1stYr'].sum(), 
    Percent_AA_to_All_1stYr= 100* x['AA_FLT_REV_1stYr'].sum() / (x['EST_AIRLN_SPEND_1stYr'].sum() if x['EST_AIRLN_SPEND_1stYr'].sum() != 0 else 1e-10))).reset_index(drop=True).drop(columns=['sum_AA', 'sum_AA_OA'])

dt2 = dt2.groupby('NPS_CATEGORY').apply(lambda x: x.assign(sum_AA=x['AA_FLT_REV'].sum(), sum_AA_OA=x['EST_AIRLN_SPEND'].sum(), 
    Percent_AA_to_All= 100* x['AA_FLT_REV'].sum() / (x['EST_AIRLN_SPEND'].sum() if x['EST_AIRLN_SPEND'].sum() != 0 else 1e-10))).reset_index(drop=True).drop(columns=['sum_AA', 'sum_AA_OA'])
dt2 = dt2.groupby('NPS_CATEGORY').apply(lambda x: x.assign(sum_AA=x['AA_FLT_REV_1stYr'].sum(), sum_AA_OA=x['EST_AIRLN_SPEND_1stYr'].sum(), 
    Percent_AA_to_All_1stYr= 100* x['AA_FLT_REV_1stYr'].sum() / (x['EST_AIRLN_SPEND_1stYr'].sum() if x['EST_AIRLN_SPEND_1stYr'].sum() != 0 else 1e-10))).reset_index(drop=True).drop(columns=['sum_AA', 'sum_AA_OA'])



dt2['AA_Spend_0'] = (dt2['AA_FLT_REV'] == 0).astype(int)
dt2['FLIGHT_REVENUE_Y1_na'] = dt2['FLIGHT_REVENUE_Y1'] 
dt2['FLIGHT_REVENUE_Y2_na'] = dt2['FLIGHT_REVENUE_Y2'] 


print("=== Issue: Why 1st Yr AA Spending =0 ? SOW data issue? ")
pd.crosstab(dt2['AA_FLT_REV_Y1_Category'], dt2['AA_FLT_REV_Category'],  margins=True, margins_name='Total')
```

```python
# Replace missing values with 0

# IMPORTANT! If the customer is in the 1st year but not in the 2nd year, then consider as churned and set values to 0. 
columns_to_fill = ['AA_FLT_REV', 'EST_AIRLN_SPEND', 'EST_AA_SOW_PERCENT', 'EP_OA_SPEND', 'AA_FLT_REV_log', 'EST_AIRLN_SPEND_log', 
                   'AA_FLT_REV_Y1', 'AA_FLT_REV_Y2', # CI table
                   'FLIGHT_REVENUE_Y1', 'FLIGHT_REVENUE_Y2', 
                   'CURRENCY_NET_REVENUE_Y1', 'CURRENCY_NET_REVENUE_Y2' ]
dt2[columns_to_fill] = dt2[columns_to_fill].fillna(0)
dt2['AA_FLT_REV_Category2'] = categorize_values(dt2['AA_FLT_REV'])

dt2['Churned'] = np.where(dt2['AA_FLT_REV_Category2'] == '0', "Churned", "Not Churned")
dt2['Diff_1to2_YoY']= dt2['AA_FLT_REV'] - dt2['AA_FLT_REV_1stYr']
dt2['TTL_Diff_1to2_YoY']= dt2['EST_AIRLN_SPEND'] - dt2['EST_AIRLN_SPEND_1stYr']


#notenote: remove those 236 customers with missing Total_ICC
```

```python
# recalculate the SOW percentage 
for col, new_col in [('AA_FLT_REV', 'Percent_AA_to_All'), ('AA_FLT_REV_1stYr', 'Percent_AA_to_All_1stYr')]:
    dt3 = dt3.groupby('NPS_CATEGORY').apply(
        lambda x: x.assign(**{
            new_col: 100 * x[col].sum() / max(x[col.replace('AA_FLT_REV', 'EST_AIRLN_SPEND')].sum(), 1e-10)
        }) ).reset_index(drop=True)
```
